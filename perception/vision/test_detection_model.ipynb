```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import numpy as np
import glob

model = YOLO("../../runs/detect/train/weights/best.pt") # YOLO12
images = sorted(glob.glob("./datasets/images-grip/sequence-2/*.jpg"))


for i in range(len(images)):
    result = model(images[i], conf=0.6)[0]
    
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(result.plot(), cv2.COLOR_BGR2RGB))
    plt.title(images[i])
    plt.axis('off')
    plt.show()
```
